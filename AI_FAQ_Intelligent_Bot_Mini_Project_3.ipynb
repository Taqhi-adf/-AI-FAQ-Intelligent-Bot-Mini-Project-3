# 8. Cosine Similarity

Most important formula in semantic search.

\cos(\theta)=\frac{A\cdot B}{||A||,||B||}

---

Interpretation:

| Score | Meaning          |
| ----- | ---------------- |
| 1     | Identical        |
| 0.9   | Very Similar     |
| 0.7   | Similar          |
| 0.5   | Somewhat Related |
| 0     | Unrelated        |
| -1    | Opposite         |

---

## Step 2

Create Knowledge Base

```python
questions = [

"What is AI?",

"What is Machine Learning?",

"What is Deep Learning?",

"What is NLP?",

"What is Generative AI?"

]
```

---

## Step 3

Answers

```python
answers = [

"AI simulates human intelligence.",

"ML learns patterns from data.",

"Deep Learning uses neural networks.",

"NLP helps machines understand language.",

"Generative AI creates content."

]
```

---

## Step 4

Load Embedding Model

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

---

## What Is This Model?

MiniLM is a pretrained transformer.

Internally:

```text
Sentence
 ↓
Transformer
 ↓
Embedding
```

Output:

384-dimensional vector.

---

## Step 5

Create FAQ Embeddings

```python
faq_embeddings = model.encode(
    questions
)
```

Shape:

```python
print(faq_embeddings.shape)
```

Output:

```text
(5,384)
```

Meaning:

```text
5 Questions

Each Question

384 Features
```

---

# Visualization

```text
Question-1
↓
[384 Numbers]

Question-2
↓
[384 Numbers]

Question-3
↓
[384 Numbers]
```

---

# Step 6

User Query

```python
query = "Explain Artificial Intelligence"
```

---

# Step 7

Query Embedding

```python
query_embedding = model.encode(
    [query]
)
```

Shape:

```text
(1,384)
```

---

# Step 8

Similarity Calculation

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(
    query_embedding,
    faq_embeddings
)
```

Output:

```text
[[0.91 0.62 0.45 0.30 0.70]]
```

---

Meaning

```text
AI = 0.91

ML = 0.62

DL = 0.45

NLP = 0.30

GenAI = 0.70
```

Highest:

```text
AI = 0.91
```

---

# Step 9

Retrieve Best Match

```python
import numpy as np

best = np.argmax(scores)

print(questions[best])
```

Output:

```text
What is AI?
```

---

# Step 10

Return Answer

```python
print(answers[best])
```

Output:

```text
AI simulates human intelligence.
```
